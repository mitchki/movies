```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
movies = pd.read_pickle('movies_all_stars.pkl')
az = pd.read_pickle('az.pkl')
```

```python
az.info()
```

```python
az['ln_gross'] = np.log(az['WorldwideGross_y'])
```

```python
p = sns.jointplot(x="ln_rate_of_return", y='ln_gross', data=az, color='coral')
p.set_axis_labels("Log Rate of Return","Log Worldwide Gross");
p;
```

```python
sns.jointplot(x='duration', y="ln_rate_of_return", data=az);
```

```python

#az = sns.load_dataset("az")

# Draw a nested boxplot to show bills by day and sex
sns.boxplot(x="content_rating", y="ln_rate_of_return",  data=az, palette="PRGn")
sns.despine(offset=10, trim=True)
```

```python
sns.boxplot(x="color", y="ln_rate_of_return",  data=az, palette="PRGn")
sns.despine(offset=10, trim=True)
```

```python
by_year = sns.boxplot(x="title_year", y="ln_rate_of_return",  data=az, palette="PRGn")
sns.despine(offset=10, trim=True)
for item in by_year.get_xticklabels():
    item.set_rotation(45)
```
